```python
import pandas as pd

# Carica il dataframe
df = pd.read_csv('beginner datasets 1\insurance.csv')

# Visualizza le dimensioni del dataset
print("Dimensioni del dataset:", df.shape)

# Anteprima del dataset
print("\nAnteprima del dataset:")
print(df.head())

# Nomi delle colonne
print("\nNomi delle colonne:")
print(df.columns)

# Medie di charges rispetto a region
print("\nMedie di charges rispetto a region:")
print(df.groupby('region')['charges'].mean())

# Medie di charges rispetto a smoker
print("\nMedie di charges rispetto a smoker:")
print(df.groupby('smoker')['charges'].mean())

# Medie di charges rispetto a sex
print("\nMedie di charges rispetto a sex:")
print(df.groupby('sex')['charges'].mean())

# Descrittori statistici di bmi
print("\nDescrittori statistici di bmi:")
print(df['bmi'].describe())


```

```python
df_iris = pd.read_csv('beginner datasets 2\iris.csv')

# Calcola la media della lunghezza dei petali di tutto il dataset
mean_petal_length = df_iris['petal_length'].mean()
print("Media della lunghezza dei petali di tutto il dataset:", mean_petal_length)

# Calcola la media della lunghezza dei petali per ogni specie di Iris, utilizzando il metodo .groupby()
mean_petal_length_by_species = df_iris.groupby('species')['petal_length'].mean()
print("\nMedia della lunghezza dei petali per ogni specie di Iris:")
print(mean_petal_length_by_species)

# Calcola media, minimo e massimo della larghezza dei sepali per ogni specie, utilizzando .groupby() e .agg()
sepal_width_stats_by_species = df_iris.groupby('species')['sepal_width'].agg(['mean', 'min', 'max'])
print("\nMedia, minimo e massimo della larghezza dei sepali per ogni specie di Iris:")
print(sepal_width_stats_by_species)
```

```python
# Carica il dataset
dataset_wine = pd.read_csv('beginner datasets 2\wine.csv')

# Media di concentrazione alcolica per ogni qualità
mean_alcohol_by_quality = dataset_wine.groupby('quality')['alcohol'].mean()
print("Media di concentrazione alcolica per ogni qualità:")
print(mean_alcohol_by_quality)

# Media totale di concentrazione alcolica
mean_alcohol_total = dataset_wine['alcohol'].mean()
print("\nMedia totale di concentrazione alcolica:", mean_alcohol_total)

# Differenza nella concentrazione alcolica per vini bianchi e vini rossi
mean_alcohol_by_type = dataset_wine.groupby('type')['alcohol'].mean()
print("\nMedia di concentrazione alcolica per vini bianchi e vini rossi:")
print(mean_alcohol_by_type)

# Analisi per il pH
# Media di pH per ogni qualità
mean_ph_by_quality = dataset_wine.groupby('quality')['pH'].mean()
print("\nMedia di pH per ogni qualità:")
print(mean_ph_by_quality)

# Media totale di pH
mean_ph_total = dataset_wine['pH'].mean()
print("\nMedia totale di pH:", mean_ph_total)

# Analisi per i solfati
# Media di solfati per ogni qualità
mean_sulfates_by_quality = dataset_wine.groupby('quality')['sulphates'].mean()
print("\nMedia di solfati per ogni qualità:")
print(mean_sulfates_by_quality)

# Media totale di solfati
mean_sulfates_total = dataset_wine['sulphates'].mean()
print("\nMedia totale di solfati:", mean_sulfates_total)
```

```python
# Carica il dataset
boston_data = pd.read_csv(r'beginner datasets 1\boston.csv')

# Media del prezzo delle case in base alla distanza dal fiume Charles (chas)
mean_medv_by_chas = boston_data.groupby('chas')['medv'].mean()
print("Media del prezzo delle case in base alla distanza dal fiume Charles (chas):")
print(mean_medv_by_chas)

# Correlazione tra tasso di criminalità (crim) e valore delle abitazioni (medv)
correlation_crim_medv = boston_data['crim'].corr(boston_data['medv'])
print("\nCorrelazione tra tasso di criminalità e valore delle abitazioni:", correlation_crim_medv)

# Media del numero di stanze (rm) rispetto al rapporto alunni-insegnanti (ptratio) e al valore delle case (medv)
mean_rm_by_ptratio = boston_data.groupby('ptratio')['rm'].mean()
mean_medv_by_ptratio = boston_data.groupby('ptratio')['medv'].mean()
print("\nMedia del numero di stanze rispetto al rapporto alunni-insegnanti:")
print(mean_rm_by_ptratio)
print("\nMedia del valore delle case rispetto al rapporto alunni-insegnanti:")
print(mean_medv_by_ptratio)

# Analisi rispetto all'accessibilità alle autostrade (rad)
# Media delle età delle abitazioni (age), del numero di stanze (rm) e delle tasse (tax)
mean_age_by_rad = boston_data.groupby('rad')['age'].mean()
mean_rm_by_rad = boston_data.groupby('rad')['rm'].mean()
mean_tax_by_rad = boston_data.groupby('rad')['tax'].mean()
print("\nMedia delle età delle abitazioni rispetto all'accessibilità alle autostrade:")
print(mean_age_by_rad)
print("\nMedia del numero di stanze rispetto all'accessibilità alle autostrade:")
print(mean_rm_by_rad)
print("\nMedia delle tasse rispetto all'accessibilità alle autostrade:")
print(mean_tax_by_rad)
```

```python
# Creiamo il DataFrame dei dipendenti
employees_df = pd.DataFrame({
    'employee_id': [101, 102, 103, 104, 105],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'department_id': [1, 2, 1, 2, 3]
})

# Creiamo il DataFrame dei dipartimenti
departments_df = pd.DataFrame({
    'department_id': [1, 2, 3],
    'department_name': ['HR', 'IT', 'Finance'],
    'location': ['New York', 'San Francisco', 'Chicago']
})

# Uniamo i DataFrame in base alla colonna comune department_id
merged_df = employees_df.merge(departments_df, on='department_id')

# Visualizziamo il DataFrame unito
print("DataFrame unito:")
print(merged_df)

# Trasformiamo la colonna department_id nell'indice per entrambi i DataFrame
employees_df.set_index('department_id', inplace=True)
departments_df.set_index('department_id', inplace=True)

# Uniamo i due DataFrame mediante il metodo .join()
joined_df = employees_df.join(departments_df, lsuffix='_employee', rsuffix='_department')

# Visualizziamo il DataFrame unito
print("\nDataFrame unito tramite il metodo .join():")
print(joined_df)
```
